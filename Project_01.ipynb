```python
## 1. Load a Csv File.
import pandas as pd
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/employees.csv")
print(data)
```

```python
## 2.  Head & Tail Function.
print(data.head(10))
print("-"*80)
print(data.tail(10))
```

```python
## 3(a). Dealing With Null Value.
print(data.isnull().sum())
```

```python
## 3(b). Deleting Null Value.
print(data.dropna())
```

```python
## 4(a). Ffill Function.
print(data.ffill())
```

```python
## 4(b). Bfill Function.
print(data.bfill())
```

```python
## 5(a). Dealing with Duplicate Value.
print(data["MANAGER_ID"].duplicated().sum())
```

```python
## 5(b). Deleting Duplicate Value.
print(data.drop_duplicates("MANAGER_ID"))
```

```python
## 6(a). Adding New Column.
data.loc[(data["SALARY"] >= 11000),"GET_BONUS"] = "No-Bonus"
data.loc[(data["SALARY"] <11000),"GET_BONUS"] = "Bonus"
print(data)
```

```python
## 6(b). Adding a new Column.
data["FULL_NAME"] = data["FIRST_NAME"] + " " + data["LAST_NAME"]
print(data)
```

```python
## 7(a). Finding Mean.
print(data["SALARY"].mean())
```

```python
## 7(b). Finding Summ.
print(data["SALARY"].sum())
```

```python
## 7(c). Max Function.
print(data["SALARY"].max())
```

```python
## 7(d). Min Function.
print(data["SALARY"].min())
```

```python
## 7(e). Info Function.
print(data.info())
```

```python
## 7(f). Describe Function.
print(data.describe())
```

```python
## 8. GRoupby Function.
# print(data)
gp = data.groupby("JOB_ID").agg({"EMPLOYEE_ID":"count"})
print(gp)
```

```python
## 9.  Pivot Table.
pt = data.pivot(index = "EMPLOYEE_ID",columns = "JOB_ID",values = "SALARY")
print(pt)
```

```python
## 10. Replacing Null Value with a Specific Value.
import numpy as np
data["MANAGER_ID"] = data["MANAGER_ID"].replace(np.nan,200)
print(data)
```
