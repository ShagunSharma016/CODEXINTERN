```python
## Bar Plot.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/expense_data_1.csv")
# print(data)
gp = data.groupby("Category")["Amount"].sum()
print(gp)
colors = ["red","green","yellow","brown","blue","violet"]
plt.bar(gp.index,gp.values,edgecolor = "black",color = colors)
plt.xlabel("Different Things")
plt.ylabel("Amount")
plt.title("Expense on different things",fontsize = 16)
plt.show()
```

```python
## Line Plot.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/expense_data_1.csv")
# print(data)
gp = data.groupby("Category")["Amount"].sum()
print(gp)
plt.plot(gp.index,gp.values,marker = "D",ls = "-.",color = "red")
plt.xlabel("Different things")
plt.ylabel("Amount")
plt.title("Expense on Different things")
plt.show()
```

```python
## Scatter Plot.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
color = np.random.randint(1,10,102)
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/expense_data_1.csv")
# print(data)
gp = data.groupby("Amount_left")["Amount"].mean()
print(gp)
plt.scatter(gp.index,gp.values,c = color,cmap = "hot",marker = "o",edgecolors = "black")
plt.xlabel("Amount Remaining")
plt.ylabel("Amount Spended")
plt.title("Home Expenses",fontsize = 16)
plt.colorbar()
plt.show()
```

```python
## Pie Chart.
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/expense_data_1.csv")
# print(data)
gp = data.groupby("Account")["Amount"].sum()
print(gp)
ex = [0,0.1,0]
plt.pie(gp.values,labels = gp.index,autopct = "%0.2f",explode = ex,shadow = True,startangle = 180,colors = ["green","silver","brown"])
plt.title("Payment Method")
plt.show()
```

```python
## Line Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data= sns.load_dataset("tips")
# print(data)
sns.lineplot(data = data,x = "day",y = "tip",hue = "sex",palette = "viridis")
plt.title("Tip by the People per Day")
plt.show()
```

```python
## Bar Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.barplot(data = data,x = "day",y = "tip",hue = "sex",palette = "GnBu",order = ["Sun","Sat","Fri","Thur"])
plt.title("Tip by person data using Bar plot")
plt.show()
```

```python
## HistoGram.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.histplot(data = data,x = "total_bill",bins = 20,hue = "sex",cumulative = False,edgecolor = "black",kde = True,palette = "viridis")
plt.title("Total_Bill Paid By the Customer Based On Gender")
plt.show()
```

```python
## Scatter Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.scatterplot(data = data,x = "total_bill",y = "tip",hue = "sex",palette = "hot",marker = "o",size = "smoker")
plt.title("Tips Based on Total_Bill")
plt.show()
```

```python
## HeatMap Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
gp = data.groupby("day",observed = False).agg({"tip":"count"})
print(gp)
sns.heatmap(gp,robust = True)
plt.title("Number of people that give tips based on day")
plt.show()
```

```python
## Count Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.countplot(data = data,x = "day",hue = "sex",palette = "viridis")
plt.title("Tips Based on Day")
plt.show()
```

```python
## Violin Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.violinplot(data = data,x = "day",y = "tip",hue = "sex",palette = "rainbow")
plt.title("Tips Data Using Violin Plot")
plt.show()
```

```python
## Pair Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.pairplot(data,hue = "sex",palette = "viridis")
plt.show()
```

```python
## Strip Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.stripplot(data = data,x = "day",y = "total_bill",hue = "sex",jitter = False,dodge = True)
plt.title("Total_Bill Per Day Using Strip Plot")
plt.show()
```

```python
## Box Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.boxplot(data = data,x = "day",y = "tip",hue = "sex",palette = "GnBu")
plt.title("Tip Shown Using Box Plot")
plt.show()
```

```python
## Cat Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.catplot(data = data,x = "day",y = "tip",hue = "sex",kind = "strip",dodge = True)
plt.title("Cat Plot to show tip per day")
plt.show()
## We can change this plot to a specific plot.
```

```python
## Relational Plot.
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
# print(data)
sns.relplot(data = data,x = "total_bill",y = "tip",hue = "sex",size = "smoker",col = "day",palette = "viridis")
plt.show()
```

```python
## Swarm Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.swarmplot(data = data,x = "day",y = "total_bill",hue = "sex",marker = "*",dodge = True)
plt.legend(bbox_to_anchor = (0.8,1.2))
plt.show()
```

```python
## Kde Plot.
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
sns.kdeplot(data = data ,x = "tip",hue = "day",palette = "viridis",multiple = "fill")
plt.title("Tip Based on Day")
plt.show()
```
